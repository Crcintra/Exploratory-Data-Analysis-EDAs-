```python
import pandas as pd

arquivo = pd.read_csv("insurance.csv")
```

```python
arquivo.head()
```

```python
##Verificando o número de linhas e colunas do arquivo

print(arquivo.shape)
```

```python
## Tirando a média de cada coluna


media = arquivo[["age", "bmi", "children", "charges"]].mean()

print(media)
```

```python
## Verificando a mediana em cada colunba
## Lembrando que a mediana éo valor que divide os dados em 2, sendo que metade dos valores estão a baixo dela e metade estão acima 

mediana = arquivo[["age", "bmi", "children", "charges"]].median()
print(mediana)
```

```python
## Verificando a moda em cada coluna

moda = arquivo[["age", "bmi", "children", "charges"]].mode()
print(moda)
```

```python
## No conjunto de dados temos a coluna charges, que representa o custo individual do plano de saúde (despesa médica anual)

##Vamos verificar se fumantes gastam mais ou menos com plano de saúde

arquivo.groupby("smoker")["charges"].mean()
```

```python
## Calculando por diferença percentual

media = arquivo.groupby("smoker")["charges"].mean()
diferenca_percentual = (media["yes"]/media["no"] - 1 ) * 100
print(f"Fumantes gastam {diferenca_percentual:.2f}% a mais.")
```

```python
## Gráfico para ver se fumantes gastam mais ou não

import matplotlib.pyplot as plt


arquivo.boxplot(column="charges", by="smoker")
plt.show()
```
